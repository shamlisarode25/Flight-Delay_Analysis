```python
# Plotting the percentage of observations that fall under each class
ax = df["DelayStatus"].value_counts().sort_values().plot(kind="barh", color=["r", "g"])
totals= []
for i in ax.patches:
    totals.append(i.get_width())
total = sum(totals)
for i in ax.patches:
     ax.text(i.get_width()+.3, i.get_y()+.20, 
     str(round((i.get_width()/total)*100, 2))+'%', 
     fontsize=10, color='black')
plt.title("DelayStatus", fontsize=20)
plt.xlabel("Count", fontsize=14)
plt.ylabel("Class", fontsize=14)
plt.show()
print(df["DelayStatus"].value_counts())
```

```python
from sklearn.utils import resample

#Separate the case of ontime, delayed
ap_ontime= df[df.DelayStatus=='Ontime']
ap_delay= df[df.DelayStatus=='Delayed']
```

```python
#Upsample the flight status
df_ontime=resample(ap_ontime,replace=True,n_samples=250000)
df_delay=resample(ap_delay,replace=True,n_samples=250000)
```

```python
#combine majority class with upsampled minority class
new_data=pd.concat([df_ontime,df_delay])
new_data.DelayStatus.value_counts()
```

```python
# Plotting the percentage of observations that fall under each class
ax = new_data["DelayStatus"].value_counts().sort_values().plot(kind="barh", color=["r", "g"])
totals= []
for i in ax.patches:
    totals.append(i.get_width())
total = sum(totals)
for i in ax.patches:
     ax.text(i.get_width()+.3, i.get_y()+.20, 
     str(round((i.get_width()/total)*100, 2))+'%', 
     fontsize=10, color='black')
plt.title("DelayStatus", fontsize=20)
plt.xlabel("Count", fontsize=14)
plt.ylabel("Class", fontsize=14)
plt.show()
print(new_data["DelayStatus"].value_counts())
```

## with sampling

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder

categorical_columns = ['Reporting_Airline','Origin','Dest']
# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Loop through each column and apply label encoding
for column in categorical_columns:
    new_data[column] = label_encoder.fit_transform(new_data[column])
```

```python
new_data.columns
```

```python
lbl = LabelEncoder()
X = new_data.drop([ 'DOT_ID_Reporting_Airline','Month',
       'IATA_CODE_Reporting_Airline','Tail_Number','OriginAirportID','OriginCityName',
        'OriginStateName', 'DestAirportID','DestCityName',
        'DestStateName','CRSDepTime', 'DepTime','DepDelay','DepDelayMinutes','CRSArrTime', 'ArrTime','ArrDelay',
       'ArrDelayMinutes','Diverted', 'CancellationCode','CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay',
       'LateAircraftDelay','Delay','DelayStatus','Flights','Quarter','DayOfWeek','Year','Cancelled','ActualElapsedTime'], axis=1)
y = new_data['DelayStatus']
y = lbl.fit_transform(y)
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                    stratify=y,
                                                    test_size=0.3,
                                                    random_state=23)
```

```python
rf = RandomForestClassifier(random_state=23)
params = {'max_features': [2,3,4,5,7]}
kfold = StratifiedKFold(n_splits=3, shuffle=True,
                        random_state=23)
gcv_rf = GridSearchCV(rf, param_grid=params, cv=kfold, scoring='neg_log_loss')
```

```python
from datetime import datetime
import time
t1 = time.time()
gcv_rf.fit(X_train, y_train)
t2 = time.time()
```

```python
print("Time taken to fit the model:", t2-t1, "seconds")
```

```python
print(gcv_rf.best_params_)
print(gcv_rf.best_score_)
best_rf = gcv_rf.best_estimator_
```

```python
best_rf = gcv_rf.best_estimator_
y_pred = best_rf.predict(X_test)
print("Accuracy with rf=",accuracy_score(y_test, y_pred))
```

```python
import seaborn as sns
from sklearn.metrics import precision_score, recall_score, classification_report, confusion_matrix
```

```python
#Method to evaluate the performance of the model
def evaluate_model(y_test,y_pred):
    print("Classification Report")
    print(classification_report(y_test, y_pred))

    print("\n---------------------------------------------\n")
    # Compute confusion matrix
    cm = confusion_matrix(y_test, y_pred)

    # Create a heatmap of the confusion matrix using Seaborn
    sns.heatmap(cm, annot=True, cmap='Greens',fmt='.0f')

    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.title('Confusion Matrix')
    plt.show()
```

```python
evaluate_model(y_test, y_pred)
```
