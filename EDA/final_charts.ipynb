```python
import matplotlib.pyplot as plt
import numpy as np
```

## Precision and Recall Bar Chart (Without Sampling)

```python
# Numbers of pairs of bars you want
N = 4

# Data on X-axis

# Specify the values of blue bars (height)
blue_bar = (0.85, 0.85, 0.83, 0.82)
# Specify the values of orange bars (height)
orange_bar = (0.95, 0.98, 1.0, 1.0)

# Position of bars on x-axis
ind = np.arange(N)

# Figure size
plt.figure(figsize=(8,5))

# Width of a bar 
width = 0.3       

# Plotting
plt.bar(ind, blue_bar , width, label='Precsion')
plt.bar(ind + width, orange_bar, width, label='Recall')

plt.xlabel('Models')
plt.title('Precision and Recall without Sampling')


# xticks()
# First argument - A list of positions at which ticks should be placed
# Second argument -  A list of labels to place at the given locations
plt.xticks(ind + width / 2, ('Random Forest', 'Catboost', 'KNN', "Logistic"))

# Finding the best position for legends and putting it
plt.legend(loc='best')
plt.show()
```

```python
##Accuracy Score (Without Sampling)
plt.figure(figsize=(5,5))
# Data for categories
categories = ['Random Forest', 'Cat Boost', 'KNN', 'Logistic Regression']

# Values for each category
values = [0.8266, 0.8365, 0.8235, 0.8242]

# Create the bar plot
plt.bar(categories, values)

# Add labels and title
plt.xlabel('Models')
plt.ylabel('Values')
plt.title('Accuracy (Without Sampling)')

# Show the plot
plt.show()
```
